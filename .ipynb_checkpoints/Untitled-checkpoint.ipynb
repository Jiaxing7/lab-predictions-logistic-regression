```python
# prep: import modules and get pwd
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import getpass  # To get the password without showing the input
password = getpass.getpass()
```

```python
# get the data
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'    #root:you can do everthing.  localhost:ip address
engine = create_engine(connection_string)
query = '''SELECT  f.film_id, c.name AS name_category, f.rental_duration, f.rental_rate, 
count(i.inventory_id) AS num_inventory, i.store_id, p.amount
FROM film f
LEFT JOIN inventory i
USING(film_id)
LEFT JOIN rental r
USING(inventory_id)
LEFT JOIN film_category
USING(film_id)
LEFT JOIN category c
USING (category_id)
LEFT JOIN payment p
USING(rental_id)
LEFT JOIN customer cu
ON (cu.customer_id = r.customer_id)
LEFT JOIN address
USING (address_id)
GROUP BY f.film_id
ORDER BY f.film_id ASC;'''

data1 = pd.read_sql_query(query, engine)
data1
```

```python
query2 = '''SELECT distinct f.film_id
FROM rental r
JOIN inventory i 
USING(inventory_id)
JOIN film f
USING(film_id)
WHERE date_format(r.rental_date,'%%y%%m') = 0505;'''

data2 = pd.read_sql_query(query2, engine)
data2
```

### 3.Read the data into a Pandas dataframe. At this point you should have 1000 rows.

```python
def  rented_in_may(row):
    if row['film_id'] in data2['film_id'].values:
        return 1
    else:
        return 0

data = pd.merge(data1, data2, on='film_id', how='left')
data['rented_in_may'] = data.apply(rented_in_may, axis=1)
data
```

### 4.Analyze extracted features (X) and transform them. You may need to encode some categorical variables, or scale numerical variables.

```python
data.isna().sum()
```

```python
data = data[data['store_id'].isna() == False]
data.isna().sum()
```

```python
# Need to X-y-split AND train-test-split BEFORE I apply transformations, 
# then train transformation on training set only
y = data['rented_in_may']
X = data.drop('rented_in_may', axis=1)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1337)
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
# from sklearn.preprocessing import StandardScaler

X_train_num = X_train.select_dtypes(include = np.number)

# Scaling data
transformer = MinMaxScaler().fit(X_train_num) # need to keep transformer
X_train_normalized = transformer.transform(X_train_num)
X_train_norm = pd.DataFrame(X_train_normalized)
```

```python
X_train_norm.columns = X_train_num.columns
X_train_norm
```

```python
X_train_cat = X_train.select_dtypes(object)
X_train_cat 
X_train_cat
```

```python

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first').fit(X_train_cat)

cols = encoder.get_feature_names(input_features=X_train_cat.columns)

X_train_cat = pd.DataFrame(encoder.transform(X_train_cat).toarray(),columns=cols)

X_train_cat

```

```python
X_train_transformed = np.concatenate([X_train_norm, X_train_cat], axis=1)
```

```python
X_train_transformed 
```

```python
from sklearn.linear_model import LogisticRegression
classification = LogisticRegression(random_state=0, solver='lbfgs',
                  multi_class='multinomial').fit(X_train_transformed, y_train)

```

```python
classification = LogisticRegression(random_state=0, solver='saga',
                  multi_class='multinomial').fit(X_train_transformed, y_train)
```

```python
# for numericals
X_test_num = X_test.select_dtypes(include = np.number)

# Scaling data
# we use the transformer that was trained on the training data
X_test_normalized = transformer.transform(X_test_num)
X_test_norm = pd.DataFrame(X_test_normalized)
X_test_norm
```

```python
X_test_cat = X_test.select_dtypes(object)
X_test_cat
```

```python
encoder = OneHotEncoder(drop='first').fit(X_test_cat)

cols = encoder.get_feature_names(input_features=X_test_cat.columns)

X_test_cat = pd.DataFrame(encoder.transform(X_test_cat).toarray(),columns=cols)

X_test_cat
```

```python
X_test_transformed = np.concatenate([X_test_norm, X_test_cat], axis=1)
X_test_transformed
```

```python
predictions = classification.predict(X_test_transformed)
classification.score(X_test_transformed, y_test)
```
